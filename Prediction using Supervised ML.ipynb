```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data=pd.read_csv('student scores.csv')
x = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values
sns.lmplot(x="Hours",y="Scores",data=data)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.2,random_state=0)
X_train=np.array(X_train)
X_train.reshape(-1,1)
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,Y_train)
prediction=lm.predict(X_test)
prediction
```

```python
plt.scatter(Y_test,prediction)
```

```python
plt.hist((Y_test,prediction),bins=50)
```

```python
sample_hours=np.array(9.25)
sample_hours=sample_hours.reshape(-1,1)
my_prediction=lm.predict(sample_hours)
print("Hours studied:",sample_hours[0][0])
print("Predicted score:",my_prediction[0])
```

```python
from sklearn import metrics
mean=metrics.mean_absolute_error(Y_test,prediction)
print("Mean absolute error:",mean)
```
